```python
def gerar_dataset():
    dados = []
    
    for page in range(1, 32):
        soup = BeautifulSoup(requests.get(f'http://sistemas.meioambiente.mg.gov.br/licenciamento/site/consulta-licenca?LicencaSearch%5Bano%5D=2019&page={page}&LicencaSearch%5Bmes%5D=6').text)
        trs = soup.find('tbody').find_all('tr')
        
        linha = []
        
        for tr in trs:
            for td in tr.find_all('td'):
                linha.append(td.text)
                
            dados.append(linha)
            linha = []
        
        
        dados.append(linha)
        
    
    df = pd.DataFrame(dados, columns=['Regional', 'Municipio', 'Empreendimento', 'CNPJ/CPF', 'Processo Administrativo', 'Nº de Protocolo', 'Modalidade', 'Classe', 'Atividade', 'Ano', 'Mes', 'Data de Publicacao', 'Decisao', 'Visualizar'])
    
    return df
```

```python
df = gerar_dataset()
```

```python
df.shape
```

```python
df = df.drop(['Processo Administrativo', 'CNPJ/CPF', 'Nº de Protocolo', 'Visualizar'], axis='columns')
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
def transformar_classe_em_numero(classe):
    return int(classe.replace('classe ', ''))
```

```python
df['Classe'] = df['Classe'].apply(lambda x: transformar_classe_em_numero(x))
```

```python
df.to_csv('processos_de_licenciamento_ambiental.csv', index=False)
```
